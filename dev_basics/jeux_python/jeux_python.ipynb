```
Jeux de casino

Le casino va lancer un dé (de façon aléatoire)
Vous devez choisir un nombre entre 1 et 6 inclus
Pour gagner, l'utilisateur doit deviner le lancé de dés
généré par le casino.
Il aura en tous 5 essais.
dès que l'utilisateur devine le nbr choisi par casino le jeux est fini
"""
```

````python
#Import de la librairie random
import random as random

#Initialisation du compte des tours de jeu
tour = 1

#Définition d'une fonction qui permet au joueur d'entrer un nombre pour jouer
def jeu_joueur():
  choix_joueur = input("Saisir un nombre entre 1 et 6 : ")
  while type(choix_joueur) != int:
    try:
      if int(choix_joueur) > 0 and int(choix_joueur) < 7:
        choix_joueur = int(choix_joueur)
        continue
      choix_joueur = input("Saisie incorrecte. Il faut choisir un nombre entre 1 et 6 : ")
    except:
      choix_joueur = input("Saisie incorrecte. Il faut choisir un entier : ")
  return choix_joueur

#Définition d'une fonction permettant de générer un message correspondant au résultat du tour
def affichage_tour(joueur, casino):
  debut = f"Choix du client {joueur}, choix du casino {casino}"
  if joueur == casino:
    fin = "Vous avez gagné"
  else:
    fin = "Vous avez perdu"
  display_message = f"{debut} \n{fin}"
  return display_message

#Initialisation de la boucle de jeu pour 5 tours
while tour <= 5:
  tentative = jeu_joueur()
  lancer_des = random.randint(1, 6)
  message_tour = affichage_tour(tentative, lancer_des)
  print(message_tour)
  tour += 1

#Affichage de la fin du jeu
print("Fin du jeu")
````

## 3. guess the number

```
- L'objectif de ce jeux est de deviner un nombre choisis par le pc
en un minimum de proposition. le nombre choisis par le pc est aléatoire et varie entre 1 et 50
- à chaque fois que l'utilisateur choisi un nombre, le pc doit afficher un message
pour lui dire s'il est suppérieur au nombre à trouver ou inférieur
- dès que l'utilisateur trouve la bonne réponse, il faudra l'informer qu'il a gagner,
arréter le jeux et afficher le nombre de tentatives



```

````python
#Import de la fonction random
import random as random

#Initialisation du compteur de tours et de variables globales
counter = 1
message_sup = "Mauvaise réponse, la bonne réponse est supérieure"
message_inf = "Mauvaise réponse, la bonne réponse est inférieure"
num_gagnant = random.randint(1, 50)

#Définition d'une fonction permettant à un joueur de saisir un nombre valide pour tenter de gagner
def tentative():
  choix_joueur = input("Choisir un nombre entre 1 et 50 : ")
  while type(choix_joueur) != int:
    try:
      if int(choix_joueur) < 1 or int(choix_joueur) > 50 :
        choix_joueur = input("Saisie invalide, vous devez saisir un entier entre 1 et 50 : ")
        continue
      choix_joueur = int(choix_joueur)
    except:
      choix_joueur = input("Saisie invalide, vous devez saisir un entier : ")
  return choix_joueur

#Définition d'une fonction permettant d'afficher un message indiquant la position du numéro gagnant ou de quitter la boucle en cas de victoire
def verif(num_j, num_g):
  if num_g > num_j:
    return message_sup
  elif num_g < num_j:
    return message_inf
  else:
    return False

#Définition d'une fonction permettant de générer un message de victoire prenant en compte le numéro gagnant et le nombre de tours
def victoire(counter_final, num_final):
  if counter_final == 1:
    affichage = f"Victoire ! Le bon numéro était le numéro {num_final} et il a été trouvé en 1 tour"
    return affichage
  else:
    affichage = f"Victoire ! Le bon numéro était le numéro {num_final} et il a été trouvé en {counter_final} tours"
    return affichage

#Première tentative
num_joueur = tentative()

#Initialisation d'une boucle qui permet au joueur de réessayer tant qu'il n'a pas réussi et qui compte le nombre de tours passés
while num_joueur != num_gagnant:
  valeure_verif = verif(num_joueur, num_gagnant)
  if valeure_verif == False:
    continue
  print(valeure_verif)
  num_joueur = tentative()
  counter += 1

#Définition d'une variable qui stocke le message de victtoire à afficher
message_final = victoire(counter, num_gagnant)

#Affichage du message de victoire
print(message_final)
````

## 4. pierre feuille ciseaux

faire un programme qui va permettre de jouer à  pierre , Feuille ciseau entre un utilisateur et un pc.

Pour rappel les régles du jeux sont :

- la pierre gagne contre le ciseau
- le ciseau gagne contre le papier
- le papier gagne contre la pierre

La partie s'arrete si un des joueurs a gagné 3 fois

voici un ex d'affichage

image.png

````python
#Import de la fonction random
import random as random

#Initialisation des variables globales
score_joueur = 0
score_ordi = 0
possibilities = {1: "pierre", 2: "papier", 3: "ciseaux", 4: "pierre", 5: "papier", 6: "ciseaux"}
#Ces deux listes servent à stocker les situations où le joueur gagne et celles où l'ordi gagne. Ma fonction de vérification additionne le carré des identifiants et ces résultats sont répertoriés
situations_joueur = [37, 20, 34]
situations_ordi = [26, 40, 25]

#Définition d'une fonction permettant à un joueur de jouer en rentrant le numéro correspondant à son choix
def jeu_joueur():
  choix_joueur = input("Saisir le numéro correspondant à votre choix : \n1. Pierre \n2. Papier \n3. Ciseaux : ")
  while type(choix_joueur) != int:
    try:
      if int(choix_joueur) < 1 or int(choix_joueur) > 3 :
        choix_joueur = input("Saisie invalide, vous devez saisir un entier entre 1 et 3 : ")
        continue
      choix_joueur = int(choix_joueur)
    except:
      choix_joueur = input("Saisie invalide, vous devez saisir un entier : ")
  return choix_joueur

#Définition d'une fonction qui permet de connaître le joueur qui remporte un point
#(en additionnant le carré des deux nombres correspondants aux choix de l'ordi et du joueur on obtient un nombre différent pour chaque situation que j'ai répertorié plus tôt )
def resultat(num_utilisateur, num_ordi):
  global score_joueur, score_ordi
  result = num_utilisateur ** 2 + num_ordi ** 2
  if result in situations_joueur:
    score_joueur += 1
    message_tour = f"Le joueur a choisi {possibilities[num_utilisateur]} et l'ordi a choisi {possibilities[num_ordi]} \nScore joueur: {score_joueur} \nScore ordi: {score_ordi} "
    return message_tour
  elif result in situations_ordi:
    score_ordi += 1
    message_tour = f"Le joueur a choisi {possibilities[num_utilisateur]} et l'ordi a choisi {possibilities[num_ordi]} \nScore joueur: {score_joueur} \nScore ordi: {score_ordi} "
    return message_tour
  else:
    message_tour = f"Le joueur a choisi {possibilities[num_utilisateur]} et l'ordi a choisi {possibilities[num_ordi]} \nScore joueur: {score_joueur} \nScore ordi: {score_ordi} "
    return message_tour

#Définition d'une fonction pour la fin du jeu
def fin_jeu(v_joueur, v_ordi):
  if v_joueur > v_ordi:
    affichage_final = f"Le joueur a gagné. Score final : Joueur {score_joueur} - Ordi {score_ordi}"
  else:
    affichage_final = f"L'ordi a gagné. Score final : Joueur {score_joueur} - Ordi {score_ordi}"
  return affichage_final

#Initialisation de la boucle qui permet de joueur au jeu
while score_joueur < 3 and score_ordi < 3:
  joueur = jeu_joueur()
  ordi = random.randint(4, 6)
  affichage_tour = resultat(joueur, ordi)
  print(affichage_tour)

#Génération d'un message de fin de jeu
message_fin_jeu = fin_jeu(score_joueur, score_ordi)

#Affichage de la fin du jeu
print(message_fin_jeu)
````

## 5  calcul mental

On souhaite créer un jeu de calcul mental dans lequel l’utilisateur devra trouver les résultats de multiplications pour deux chiffres (entre 1 et 9) posées au hasard par l’ordinateur. Dès que le joueur aura obtenu 5 bonnes réponses, le jeu s’arretera et affichera la durée du jeu.
 ci dessous un exemple d'affichage

image.png

````python
#Import de la fonction random et de la librairie time
import random as random
import time

#Déclaration de variable globale
vraies_reponses = 0

#Définition d'une fonction permettant à un joueur de saisir un nombre valide pour tenter de gagner
def reponse_joueur():
  choix_joueur = input("Réponse : ")
  while type(choix_joueur) != int:
    try:
      if int(choix_joueur) < 1 or int(choix_joueur) > 81 :
        choix_joueur = input("Saisie invalide, vous devez saisir un entier entre 1 et 81 : ")
        continue
      choix_joueur = int(choix_joueur)
    except:
      choix_joueur = input("Saisie invalide, vous devez saisir un entier : ")
  return choix_joueur

#Définition d'une fonction qui vérifie la réponse du joueur
def verification(bonne_reponse, tentative_reponse):
  global vraies_reponses
  if bonne_reponse == tentative_reponse:
    vraies_reponses += 1
    message_rep = f"Correct, le nombre actuel de bonnes réponses est {vraies_reponses}"
    return message_rep
  else:
    message_rep = f"Mauvaise réponse, le nombre actuel de bonnes réponses est {vraies_reponses}"
    return message_rep

#Lancement du timer
debut = time.time()

#Lancement de la boucle de jeu
while vraies_reponses < 5:
  num1 = random.randint(1, 9)
  num2 = random.randint(1, 9)
  la_reponse = num1 * num2
  print(f"{num1} X {num2} :")
  rep = reponse_joueur()
  affiche_tour = verification(la_reponse, rep)
  print(affiche_tour)

#Fin du timer
fin = time.time()

#Calcul du temps de jeu
temps = fin - debut
temps = round(temps, 2)

#Affichage de la fin du jeu
print("Le jeu est fini")
print(f"Temps = {temps} secondes")
````
